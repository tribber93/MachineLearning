```python
import numpy as np
import pandas as pd
```

```python
path = '/content/drive/MyDrive/Machine Learning/dataset/dataset_uts-2021.csv'
df = pd.read_csv(path)
df
```

```python
df.dtypes
```

```python
# a. Menghapus baris yang hilang
df.dropna(inplace = True)
df
```

```python
# b. Menghapus kolom yang berisi data yang tidak relevan (Hapus Kolom No)
df.drop(["No"],inplace=True,axis=1)
df
```

```python
# c. Perbaiki kesalahan input (intenal harusnya Internal)
df.loc[4,'Project'] = "Internal"
df
```

```python
# d. Ubah Tipe data float64 untuk Mobile Phone menjadi Teks atau Object

# convert ke int untuk menghapus .0 lalu baru ke string
df['Mobile Phone'] = df['Mobile Phone'].astype(int).astype(str) 
df
```

```python
df.dtypes
```

```python
# e. Ganti nama semua kolom dalam bahasa Indonesia
df = df.rename(columns={
    'Name': 'Nama', 
    'Address':'Alamat', 
    'Mobile Phone':'Nomor Handphone',
    'Project': 'Projek'
}) 
df
```

```python
# f. Simpan dalam file dataset yang baru
path2 = '/content/drive/MyDrive/Machine Learning/dataset/dataset-uts-baru-2022.csv'
df.to_csv(path2)
```
