```python
import pandas as pd
import numpy as np
import random
from sklearn.model_selection import train_test_split
```

```python
path ='/content/drive/MyDrive/Machine Learning/dataset/Indonesian Salary by Region (1997-2022).csv'
df = pd.read_csv(path)
df
```

```python
df.dtypes
```

# 1. Menggunakan Random

Pembagian 70% dan 30%

```python
# menentukan nilai random untuk menentukan index ke berapa yang masuk ke data train dan data test
""" 
  rd menampung nilai array random antara 1 dan 2 sebanyak panjang kolom region
  dengan pembagian(p) 70% dan 30%
"""
rd = np.random.choice([1,2], size=(len(df['REGION'])), p=[0.7,0.3])
rd
```

```python
train = df.loc[rd==1,] # mengambil data train dengan index yang memiliki nilai rd = 1
train
```

```python
test = df.loc[rd==2,] # mengambil data test dengan index yang memiliki nilai rd = 2
test
```

```python
print('-------- Pembagian 70% dan 30% --------\n')
print(f'jumlah data train -> {len(train)}')
print(f'jumlah data test -> {len(test)}')
```

Pembagian 80% dan 20%

```python
rd = np.random.choice([1,2], size=(len(df['REGION'])), p=[0.8,0.2])
rd
```

```python
train = df.loc[rd==1,]
train
```

```python
test = df.loc[rd==2,]
test
```

```python
print('-------- Pembagian 80% dan 20% --------\n')
print(f'jumlah data train -> {len(train)}')
print(f'jumlah data test -> {len(test)}')
```

# 2. Menggunakan train_test_split

```python
X = df[['REGION','YEAR']] #variable X untuk kolom wilayah dan tahun
y = df['SALARY'] #variable y untuk kolom gaji
```

```python
X
```

```python
y
```

Pembagian 70% dan 30%

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7)
```

```python
print("-------- Pembagian 70% dan 30% --------\n")
print(f'Jumlah X_train -> {len(X_train)}')
print(f'Jumlah X_test -> {len(X_test)}')
print(f'Jumlah y_train -> {len(y_train)}')
print(f'Jumlah y_test -> {len(y_test)}')
```

Pembagian 80% dan 20%

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
print("-------- Pembagian 80% dan 20% --------\n")
print(f'Jumlah X_train -> {len(X_train)}')
print(f'Jumlah X_test -> {len(X_test)}')
print(f'Jumlah y_train -> {len(y_train)}')
print(f'Jumlah y_test -> {len(y_test)}')
```
